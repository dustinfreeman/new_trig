```python
from new_trig import *
import matplotlib.pyplot as plt
```

```python
#generating plot range
PLT_PRECISION = 100
plt_range = []
for i in range(-PLT_PRECISION, PLT_PRECISION):
	plt_range.append(2.0*pi*i/(PLT_PRECISION))
```

```python
def graph_fun(fun, fun_name):
	y = []
	for x in plt_range:
		y.append(fun(x))

	plt.ylim(-2*pi,2*pi)
	plt.xlim(-2*pi,2*pi)

	plt.plot(plt_range, y)

	plt.title(fun_name)
	plt.grid(True)
	plt.show()
```

```python
#the old
graph_fun(sin, 'sin')
graph_fun(cos, 'cos')
graph_fun(tan, 'tan')


#the new
graph_fun(phyxyx, 'phyxyx')
graph_fun(negtan, 'negtan')
graph_fun(gasmin, 'gasmin')
graph_fun(tofsin, 'tofsin')
graph_fun(drin, 'drin')
graph_fun(codrin, 'codrin')
graph_fun(cocosine, 'cocosine')
graph_fun(cosvnx, 'cosvnx')
```
